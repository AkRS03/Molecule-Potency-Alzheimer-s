```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Compiled_descriptors_2.csv')
```

```python
df=df.iloc[:,1:]
```

```python
sns.distplot(df['MATS2p'],hist=False)
```

```python
mask1=df['MATS2p']<=-0.15
mask2=df['MATS2p']>=0.15
```

```python
abs(df[mask1|mask2].corr()['Class']).sort_values(ascending=False)
```

```python
df[mask1|mask2]
```

```python
m=[]
for ele in df['MATS2p']:
    if abs(ele)>=0.1:
        m.append(1)
   
    
    else:
        m.append(0)

df['MATS2p_Class']=m
```
