```python
df_final.drop_duplicates(inplace=True,)
```

```python
df_final=df_final.T
```

```python
df_final.drop(['SMILES','CASRN','MOLECULEID','ARTICLEID'],inplace=True,axis=1)
```

```python
df_final
```

```python
m = [0 if ele <= 1900 else 1 if ele <= 5800 else 2 for ele in df_final['IC50']]
```

```python
df_final['Class']=m
df_final.drop('IC50', axis=1, inplace=True)
```

```python
l = []
for i in range(0, 2280):
    for j in range(i+1, len(df_final.columns.to_list())):
        correlation = df_final[df_final.columns[i]].corr(df_final[df_final.columns[j]])
        if correlation >= 0.95:
            l.append(df_final.columns[j])
        else:
            pass
```

```python
j=[]
for ele in l:
    if ele not in j:
        j.append(ele)
```

```python
df_final.drop(j,axis=1,inplace=True)
```

```python
df_final
```

```python
k=[ele for ele in df_final.columns if df_final[ele].nunique()<=4 if df_final['Class'].corr(df_final[ele]<=0.20)]
```

```python
k.remove('Class')
```

```python
len(k)
```

```python
df_final.drop(k,axis=1,inplace=True)
```
